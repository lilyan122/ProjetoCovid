**Carregando dados de mortos e recuperados**

```python
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")

display(deaths.head())
display(recovered.head())
```

```python
def latest_by_country(data):
  return data.groupby("Country/Region").sum().iloc[:,-1]

latest_by_country(confirmed).head()
```

Para cada país o número, no último dia disponível dos dados, de confirmados, mortes e recuperados.

```python
informations = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]
combined = pd.concat(informations, axis=1)
combined.columns = ["confirmed", "deaths", "recovered"]
combined.head()
```

```python
sum_up = combined.sum()
letality_rate_1 = sum_up["deaths"] / sum_up["confirmed"]
print(letality_rate_1 * 100)
```

1,07% das pessoas que são casos confirmados no conjunto de dados que nós temos acesso morreram.

```python
letality_rate_1 = combined["deaths"] / combined["confirmed"] * 100
combined["letality_rate_1"] = letality_rate_1
combined.head()
```

Letalidade por país

```python
sum_up = combined.sum()
letality_rate_1 = sum_up["deaths"] / sum_up["confirmed"]
letality_rate_2 = sum_up["deaths"] / (sum_up["deaths"] + sum_up["recovered"])
print(letality_rate_1 * 100)
print(letality_rate_2 * 100)
```

A segunda estimativa indica que dos casos em que foi terminada a análise, em que o paciente foi enviado de volta para casa porque resolveu ou faleceu, a taxa de mortalidade nesse grupo foi de 100,0%

```python

letality_rate_1 = combined["deaths"] / combined["confirmed"] * 100
letality_rate_2 = combined["deaths"] / (combined["recovered"] + combined["deaths"]) * 100
combined["letality_rate_1"] = letality_rate_1
combined["letality_rate_2"] = letality_rate_2
combined.head()
```

```python
combined.sort_values("letality_rate_1", ascending=False).head(10)
```

Ordenar pela letalidade

```python
#  Visualização horrenda
combined[["confirmed","letality_rate_1"]].plot()
```

```python
import numpy as np

confirmed["china_label"]= np.where(confirmed["is_china"],'China', 'others')
display(confirmed.head())
display(confirmed.tail())
```

```python
confirmed.groupby("china_label").sum()['9/13/22'].plot(kind='bar')
plt.title(f"Casos confirmados acumulados até {'9/13/22'}")
plt.show()
```

```python

summed = confirmed.groupby("china_label").sum()
summed['delta'] = summed['9/13/22'] - summed['9/12/22']
summed['delta'].plot(kind='bar')
plt.title(f"Novos casos confirmados em {'9/13/22'}")
plt.show()
```

```python
summed.T
```

```python
summed.T[2:-2].plot()
plt.title('Desenvolvimento de casos confirmados com passar do tempo')
plt.show()
```

```python
differences = summed.T[2:-2].diff().dropna()
differences.plot()
plt.title("Casos novos com decorrer do tempo")
plt.show()
```

```python
combined_expanded = combined.join(un_population_total).sort_values("confirmed", ascending=False).dropna()
combined_expanded.head(10)
```

```python
#Calculando agora a nossa taxa de incidência.
combined_expanded['incidence_ratio'] = combined_expanded['confirmed'] / combined_expanded['PopTotal'] * 100000
combined_expanded.head(10)
```

```python
#Calculando agora a nossa taxa de Mortalidade
combined_expanded['mortality_rate'] = combined_expanded['deaths'] / combined_expanded['PopTotal'] * 100000
combined_expanded.head(10)
```
